## Estructuras de datos

Como su nombre lo indica, son estructuras caracterizadas por contener el mismo o diferentes tipos de datos. 

### Arreglo
Un arreglo es una colección de datos del mismo tipo con un número arbitrario de dimensiones.

<b>Nota:</b> A lo largo del código se importan varias librerías; una librería es un conjunto de funciones que permiten realizar diferentes operaciones, ahorrando tiempo y esfuerzo.

[**NumPy**](https://numpy.org/doc/stable/)

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3])
print(arr1)
print(arr1.shape)
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2)
print(arr2.shape)
```

### Diccionario
Un diccionario es una colección no ordenada de parejas compuestas de llaves y valores, donde las llaves son únicas y los valores pueden ser de cualquier tipo. 

```python
dicc1 = {'nombre':'Juan', 'edad':40, 'estado_civil':'Casado'}
print(dicc1)
```

```python
print(dicc1.keys())
print(dicc1.values())
```

### Dataframe
Un DataFrame es una estructura de datos de dos dimensiones que puede contener datos de diferentes tipos (similar a una tabla). Esta colección de datos suele ser la más utilizada para realizar análisis de información.

[**pandas**](https://pandas.pydata.org/)

```python
import pandas as pd
```

```python
dicc2 = {'nombre':['Juan','Andrea','David'], 'edad':[40,34,26], 'estado_civil':['Casado','Soltero','Casado']}
print(dicc2)
```

```python
df = pd.DataFrame(dicc2)
df
```

## Estructuras de control
Como su nombre lo indica, las estructuras de control permiten controlar el orden en que se ejecuta el código. 

Las estructuras de control más utilizadas son:


| Estructura de control | Descripción       |
| --------- | --------------- |
|    if, elif, else    | Evalua una condición y actua en consecuencia |
|    for    | Ejecuta un ciclo por un número específico de veces |
|    while      | Ejecuta un ciclo mientras se cumpla una condición |

### Estructuras condicionales

Las estructuras condicionales permiten ejecutar diferentes operaciones dependiendo de una condición.

#### if

La sentencia <b>if</b> realiza una acción cuando se cumple una condición, y cuando no se cumple se ignora.

```python
# Si la condición es verdadera (True) se realiza la operación

if 4 > 2:
    print('Se cumple la condición')
```

```python
# Si no se cumple la condición, no se realiza la operación

if 4 < 2:
    print('Se cumple la condición')
```

#### If...else

La sentencia <b>else</b> se utiliza para indicar la operación que se realiza si la condición de un <b>if</b> no se cumple.

```python
# La estructura else indica la operación que se realiza si la condición de un if no se cumple (False)

if 4 < 2:
    print('Se cumple la condición')

else:
    print('No se cumple la condición del if')
```

#### If...elif...else

Se pueden definir múltiples condiciones combinando <b>if</b>, <b>elif</b> y <b>else</b>, donde dichas condiciones son mutuamente excluyentes.

```python
a = 2
if a == 1:
    print('Se cumple la condición 1')
elif a == 2:
    print('Se cumple la condición 2')
elif a == 3:
    print('Se cumple la condición 3')
else:
    print('No se cumple ninguna condición')
```

### Estructuras iterativas

Las estructuras iterativas (ciclos) son métodos que permiten repetir un bloque de instrucciones.

#### for

La sentencia <b>for</b> permite realizar una acción para cada elemento de un conjunto de elementos, la cual es controlada por un contador que se incrementa en cada iteración.

```python
for itn in [1,2,3,4,5]:
    print(itn)
```

```python
for itn in [1,2,3,4,5]:
    print(itn * 2)
```

```python
for itn in range(5):
    print(itn)
```

#### while

La sentencia <b>while</b> realiza una acción mientras que una condición se mantenga como verdadera.

```python
b = 1

while b < 5:
    print(b)
    b += 1
```
